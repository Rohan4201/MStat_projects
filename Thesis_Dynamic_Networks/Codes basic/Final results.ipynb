## Theoretical Final expected edge density

```python
final_matrix = symmetric_matrices[-1]
transition_matrix = np.array([[0.8, 0.05, 0.15], [0.2, 0.75, 0.05], [0.25, 0.2, 0.55]])
transition_matrix_power = np.linalg.matrix_power(transition_matrix, 499)

initial_state_matrix = np.array([1*(initial_states==0), 1*(initial_states == 1), 1*(initial_states ==2)]).T
```

```python
first_term_expected_density = np.dot(initial_state_matrix, transition_matrix_power)
```

```python
final_time_matrix = np.dot(np.dot(first_term_expected_density, final_matrix), initial_state_matrix.T)
num_nodes = final_time_matrix.shape[0]


np.sum(np.tril(final_time_matrix, k = -1))/(num_nodes * (num_nodes-1) * 0.5)
```
